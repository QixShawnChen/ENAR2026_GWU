```python
import pandas as pd
import numpy as np
```

```python
filepath = "rsid_unique_variants.csv"
df = pd.read_csv(
   filepath,
    sep="\t"
)

df
```

```python
df5 = pd.read_csv(
    "data_mutations.txt",
    sep="\t",
    comment="#"
)

df5
```

```python
df_inner = pd.merge(
    df,
    df5,
    on=["Chromosome", "Start_Position", "End_Position"],
    how="inner"
)
df_inner
```

```python
df_unique = df_inner.drop_duplicates()
df_unique
```

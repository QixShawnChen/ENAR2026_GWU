```python
df_new = df[["›#Uploaded_variation", "Location", "Existing_variation", "SYMBOL",
    "Gene",
    "SYMBOL",
    "Consequence",
    "IMPACT",
    "CLIN_SIG",]]
```

```python
df_new["rsid"] = df["Existing_variation"].str.extract(r"(rs\d+)")
df_new["clinvar"] = df["Existing_variation"].str.extract(r"(CM\d+)")
```

```python
df_new
```

```python
df_new.loc[df_new["rsid"] != ".", "rsid"].nunique()
```

```python
tmp = df_new["Location"].str.split(":", expand=True)
pos = tmp[1].str.split("-", expand=True)

df_new["Chromosome"] = tmp[0]
df_new["Start_Position"] = pos[0].astype(int)
df_new["End_Position"] = pos[1].astype(int)

df_new["Allele"] = (
    df_new["›#Uploaded_variation"]
    .str.split("_", expand=True)[2]
)
```

```python
df_new
```

```python
# Keep the cols with the following values in functional_consequences function.
functional_consequences = {
    "missense_variant",
    "stop_gained",
    "stop_lost",
    "frameshift_variant",
    "splice_acceptor_variant",
    "splice_donor_variant",
    "start_lost",
    "inframe_insertion",
    "inframe_deletion",
}
```

```python
# Keep the pathogenic variants with HIGH or at least MODERATE impact.
df_filtered = df_new[
    df_new["IMPACT"].isin(["HIGH", "MODERATE"])
    &
    df_new["Consequence"].apply(
        lambda x: any(c in x for c in functional_consequences)
        if isinstance(x, str) else False
    )

    &
    (df_new["CLIN_SIG"].str.contains("pathogenic", case=False, na=False))
].copy()
```

```python
df_filtered
```

```python
df_filtered.loc[df_filtered["rsid"] != ".", "rsid"].nunique()
```

```python
# Remove Duplicates
df_filtered_unique = df_filtered.drop_duplicates(
    subset=["Chromosome", "Start_Position", "End_Position", "Allele", "rsid", "CLIN_SIG", "IMPACT", "Consequence"]
)
```

```python
df_filtered_unique
```

```python
df_filtered_unique.to_csv(
    "rsid_unique_variants2.csv",
    sep="\t",
    index=False
)
```

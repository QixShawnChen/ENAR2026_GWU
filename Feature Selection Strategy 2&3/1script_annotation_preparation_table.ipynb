```python
import pandas as pd
import numpy as np
```

```python
filepath = "df5_data_mutation.csv"
```

```python
df1 = pd.read_csv("df5_data_mutation.csv")
```

```python
print(df1.columns.tolist())
```

```python
df1_selected = df1[["Chromosome", "Start_Position", "End_Position", "Reference_Allele", "Tumor_Seq_Allele2"]].copy()
```

```python
df1_selected
```

```python
df1_selected["allele"] = (
    df1_selected["Reference_Allele"].astype(str)
    + "/"
    + df1_selected["Tumor_Seq_Allele2"].astype(str)
)
```

```python
df1_selected
```

```python
df1_selected["length"] = (
    df1_selected["End_Position"]
    - df1_selected["Start_Position"]
    + 1
)
```

```python
df1_selected
```

```python
# For 'Ensembl', it requires the format: "Chromosome Starting_Pos End_Pos allele_pair length".
df_out = df1_selected[
    [
        "Chromosome",
        "Start_Position",
        "End_Position",
        "allele",
        "length",
    ]
]
```

```python
df_out
```

```python
df_out.to_csv(
    "variants_for_design.txt",
    sep=" ",
    index=False,
    header=False
)
```

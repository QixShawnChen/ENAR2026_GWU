```python
import pandas as pd

df1 = pd.read_csv(
    "data_clinical_patient.txt",
    sep="\t",
    comment="#"
)

df2 = pd.read_csv(
    "data_clinical_sample.txt",
    sep="\t",
    comment="#"
)

df3 = pd.read_csv(
    "data_methylation_promoters_rrbs.txt",
    sep="\t"
)

df4 = pd.read_csv(
    "data_mrna_illumina_microarray.txt",
    sep="\t"
)

df5 = pd.read_csv(
    "data_mutations.txt",
    sep="\t",
    comment="#"
)
```

```python
df3_T = df3.transpose()


df3_T.columns = df3_T.loc["Hugo_Symbol"]
df3_T = df3_T.drop(index="Hugo_Symbol")
df3_T = df3_T.reset_index().rename(columns={"index": "PATIENT_ID"})
df3_T = df3_T.rename(columns={"Hugo_Symbol": ""})
df3_T = df3_T.rename(
    columns=lambda c: c if c == "PATIENT_ID" else f"{c}_methy"
)
```

```python
df2 = df2.drop(columns=["SAMPLE_ID"])
df4 = df4.drop(columns=["Entrez_Gene_Id"])
df4_T = df4.transpose()
```

```python
df4_T.columns = df4_T.loc["Hugo_Symbol"]
df4_T = df4_T.drop(index="Hugo_Symbol")
df4_T = df4_T.reset_index().rename(columns={"index": "PATIENT_ID"})
df4_T = df4_T.rename(columns={"Hugo_Symbol": ""})

df4_T = df4_T.rename(
    columns=lambda c: c if c == "PATIENT_ID" else f"{c}_mrna"
)
```
